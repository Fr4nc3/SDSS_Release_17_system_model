# I. Data Collection and Preprocessing

```python
import pandas as pd
file_path = 'photometric_stars.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.describe(include='all')
```

```python
data[['ra', 'dec', 'u', 'g', 'r', 'i', 'z']].describe()
```

```python
problematic_flags = ['EDGE', 'NOPROFILE', 'SATURATED', 'COSMIC_RAY', 'INTERP_CENTER']

available_flags = [flag for flag in problematic_flags if flag in data.columns]

filtered_data = data.copy()
for flag in available_flags:
    if flag == 'INTERP_CENTER' and 'COSMIC_RAY' in available_flags:
        filtered_data = filtered_data[
            ~((filtered_data[flag]) & (filtered_data['COSMIC_RAY']))
        ]
    else:
        filtered_data = filtered_data[~filtered_data[flag]]

error_columns = [col for col in data.columns if 'psfMagError' in col]
filtered_data = filtered_data[
    (filtered_data[error_columns] <= 0.2).all(axis=1)
]

original_count = data.shape[0]
filtered_count = filtered_data.shape[0]
excluded_count = original_count - filtered_count

print(f"Original dataset size: {original_count}")
print(f"Filtered dataset size: {filtered_count}")
print(f"Number of rows excluded: {excluded_count}")
print(f"Percentage of data retained: {(filtered_count / original_count) * 100:.2f}%")
```

```python

error_columns = ['psfMagErr_u', 'psfMagErr_g', 'psfMagErr_r', 'psfMagErr_i', 'psfMagErr_z']


filtered_data = data[
    (data[error_columns] <= 0.5).all(axis=1)  # Retain rows where all errors are <= 0.5
]


original_count = data.shape[0]
filtered_count = filtered_data.shape[0]
excluded_count = original_count - filtered_count

print(f"Original dataset size: {original_count}")
print(f"Filtered dataset size: {filtered_count}")
print(f"Number of rows excluded: {excluded_count}")
print(f"Percentage of data retained: {(filtered_count / original_count) * 100:.2f}%")
```

```python

flags_to_check = ['EDGE', 'COSMIC_RAY', 'NOPROFILE']  # Replace with actual flag column names
flag_summary = {flag: data[flag].sum() for flag in flags_to_check if flag in data.columns}

error_columns = ['psfMagErr_u', 'psfMagErr_g', 'psfMagErr_r', 'psfMagErr_i', 'psfMagErr_z']
psf_error_summary = data[error_columns].describe()

original_count = data.shape[0]
filtered_count = filtered_data.shape[0]
retention_rate = (filtered_count / original_count) * 100

magnitude_columns = ['psfMag_u', 'psfMag_g', 'psfMag_r', 'psfMag_i', 'psfMag_z']
magnitude_summary = data[magnitude_columns].describe()


data['g-r'] = data['psfMag_g'] - data['psfMag_r']
data['r-i'] = data['psfMag_r'] - data['psfMag_i']
color_summary = data[['g-r', 'r-i']].describe()


print("Flag Summary:")
print(flag_summary)

print("\nPSF Error Summary:")
print(psf_error_summary)

print("\nRetention Rate:")
print(f"Original dataset size: {original_count}")
print(f"Filtered dataset size: {filtered_count}")
print(f"Percentage of data retained: {retention_rate:.2f}%")

print("\nMagnitude Summary:")
print(magnitude_summary)

print("\nColor Index Summary:")
print(color_summary)
```

# II. Workflow

```python
ra_range = (data['ra'].min(), data['ra'].max())
dec_range = (data['dec'].min(), data['dec'].max())

magnitude_columns = ['psfMag_u', 'psfMag_g', 'psfMag_r', 'psfMag_i', 'psfMag_z']
magnitude_summary = data[magnitude_columns].describe()

error_columns = ['psfMagErr_u', 'psfMagErr_g', 'psfMagErr_r', 'psfMagErr_i', 'psfMagErr_z']
psf_error_summary = data[error_columns].describe()

flag_columns = ['EDGE', 'COSMIC_RAY', 'NOPROFILE']
flag_counts = {flag: data[flag].sum() for flag in flag_columns if flag in data.columns}

total_rows = data.shape[0]
total_columns = data.shape[1]
critical_columns_present = all(col in data.columns for col in ['ra', 'dec', *magnitude_columns])


print(f"RA Range: {ra_range}")
print(f"Dec Range: {dec_range}")
print("\nPhotometric Magnitudes Summary:")
print(magnitude_summary)

print("\nPSF Magnitude Errors Summary:")
print(psf_error_summary)

print("\nQuality Flags Exclusion Counts:")
print(flag_counts)

print(f"\nTotal Rows: {total_rows}")
print(f"Total Columns: {total_columns}")
print(f"Critical Columns Present: {critical_columns_present}")
```

```python
import matplotlib.pyplot as plt

data['g-r'] = data['psfMag_g'] - data['psfMag_r']
data['r-i'] = data['psfMag_r'] - data['psfMag_i']


plt.figure(figsize=(8, 6))
plt.scatter(data['g-r'], data['psfMag_r'], s=1, alpha=0.5)
plt.gca().invert_yaxis()  # Invert y-axis for magnitude
plt.title('Color-Magnitude Diagram (g-r vs. r)')
plt.xlabel('g-r')
plt.ylabel('r (magnitude)')
plt.grid()
plt.show()


magnitude_columns = ['psfMag_u', 'psfMag_g', 'psfMag_r', 'psfMag_i', 'psfMag_z']
for mag in magnitude_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(data[mag], bins=50, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f'Distribution of {mag}')
    plt.xlabel('Magnitude')
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(data['ra'], data['dec'], s=1, alpha=0.5)
plt.title('Spatial Distribution of Stars')
plt.xlabel('RA (degrees)')
plt.ylabel('Dec (degrees)')
plt.grid()
plt.show()


color_indices = ['g-r', 'r-i']
for color in color_indices:
    plt.figure(figsize=(8, 6))
    plt.hist(data[color], bins=50, alpha=0.7, color='green', edgecolor='black')
    plt.title(f'Distribution of {color}')
    plt.xlabel('Color Index')
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import os

# Prevent joblib warnings
os.environ["LOKY_MAX_CPU_COUNT"] = "1"

# Prepare data for clustering
data['g-r'] = data['psfMag_g'] - data['psfMag_r']
data['r-i'] = data['psfMag_r'] - data['psfMag_i']
features = data[['g-r', 'r-i']].dropna()

# Normalize features
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features)

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans_labels = kmeans.fit_predict(normalized_features)

# Visualize K-Means Clustering
plt.scatter(features['g-r'], features['r-i'], c=kmeans_labels, cmap='viridis', s=10)
plt.xlabel('g-r')
plt.ylabel('r-i')
plt.title('K-Means Clustering')
plt.show()

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=10)
dbscan_labels = dbscan.fit_predict(normalized_features)

# Visualize DBSCAN Clustering
plt.scatter(features['g-r'], features['r-i'], c=dbscan_labels, cmap='viridis', s=10)
plt.xlabel('g-r')
plt.ylabel('r-i')
plt.title('DBSCAN Clustering')
plt.show()
```
