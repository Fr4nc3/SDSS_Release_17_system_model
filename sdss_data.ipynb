```python
import pandas as pd
file_path = 'photometric_stars.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
problematic_flags = ['EDGE', 'NOPROFILE', 'SATURATED', 'COSMIC_RAY', 'INTERP_CENTER']

available_flags = [flag for flag in problematic_flags if flag in data.columns]

filtered_data = data.copy()
for flag in available_flags:
    if flag == 'INTERP_CENTER' and 'COSMIC_RAY' in available_flags:
        filtered_data = filtered_data[
            ~((filtered_data[flag]) & (filtered_data['COSMIC_RAY']))
        ]
    else:
        filtered_data = filtered_data[~filtered_data[flag]]

error_columns = [col for col in data.columns if 'psfMagError' in col]
filtered_data = filtered_data[
    (filtered_data[error_columns] <= 0.2).all(axis=1)
]

original_count = data.shape[0]
filtered_count = filtered_data.shape[0]
excluded_count = original_count - filtered_count

print(f"Original dataset size: {original_count}")
print(f"Filtered dataset size: {filtered_count}")
print(f"Number of rows excluded: {excluded_count}")
print(f"Percentage of data retained: {(filtered_count / original_count) * 100:.2f}%")
```

# Integration with Clustering and Classification

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
import numpy as np


np.random.seed(42)
data = pd.DataFrame({
    'g-r': np.concatenate([np.random.normal(0.5, 0.1, 100), np.random.normal(1.5, 0.1, 100)]),
    'r-i': np.concatenate([np.random.normal(0.3, 0.05, 100), np.random.normal(0.7, 0.05, 100)])
})

# K-Means clustering on 'g-r' and 'r-i'
kmeans = KMeans(n_clusters=2, random_state=42)
data['cluster'] = kmeans.fit_predict(data[['g-r', 'r-i']])

# Calculate silhouette score
silhouette_avg = silhouette_score(data[['g-r', 'r-i']], data['cluster'])

# Cluster centers
cluster_centers = kmeans.cluster_centers_

# Output results
silhouette_avg, cluster_centers
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Simulated data for RA and Dec within given ranges
np.random.seed(42)
ra = np.random.uniform(180.0005, 180.9999, 5000)
dec = np.random.uniform(-0.4999, 0.4201, 5000)

# Calculate density using Gaussian KDE
positions = np.vstack([ra, dec])
kde = gaussian_kde(positions)(positions)

# Generate a grid for density map
x_grid, y_grid = np.linspace(ra.min(), ra.max(), 100), np.linspace(dec.min(), dec.max(), 100)
X, Y = np.meshgrid(x_grid, y_grid)
positions_grid = np.vstack([X.ravel(), Y.ravel()])
density_grid = gaussian_kde(positions)(positions_grid).reshape(X.shape)

# Scatter plot with density overlay
plt.figure(figsize=(10, 8))
plt.scatter(ra, dec, c=kde, cmap='viridis', s=10, label='Star Density')
plt.colorbar(label='Density')
plt.title('Scatter Plot of RA vs. Dec with Density Overlay')
plt.xlabel('RA (degrees)')
plt.ylabel('Dec (degrees)')
plt.legend()
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(kde, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Density Histogram')
plt.xlabel('Density')
plt.ylabel('Frequency')
plt.show()

# Calculate grid-based density
grid_counts = np.histogram2d(ra, dec, bins=(20, 20))[0]

# Display grid-based density map
plt.figure(figsize=(10, 8))
plt.imshow(grid_counts.T, extent=[ra.min(), ra.max(), dec.min(), dec.max()], origin='lower', cmap='plasma', aspect='auto')
plt.colorbar(label='Grid-Based Density')
plt.title('Grid-Based Density Map of RA vs. Dec')
plt.xlabel('RA (degrees)')
plt.ylabel('Dec (degrees)')
plt.show()

# Summary statistics for density
density_summary = {
    "Mean Density": kde.mean(),
    "Median Density": np.median(kde),
    "Density Range": (kde.min(), kde.max()),
    "Grid Density Range": (grid_counts.min(), grid_counts.max()),
}

density_summary
```
